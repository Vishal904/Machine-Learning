```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('Train/Jamie_Before.jpg')
image =  cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

goggle = cv2.imread('Train/glasses.png',-1)

mustache = cv2.imread('Train/mustache.png',-1)



eyes_detector = cv2.CascadeClassifier('Train/third-party/frontalEyes35x16.xml') 
eyes = eyes_detector.detectMultiScale(image) 

nose_detector = cv2.CascadeClassifier('Train/third-party/Nose18x15.xml')
nose = nose_detector.detectMultiScale(image,1.5,5)



width = int(goggle.shape[1] * 85 / 100)
height = int(goggle.shape[0] * 90 / 100)
dim = (width, height)
goggle = cv2.resize(goggle, dim, interpolation = cv2.INTER_AREA)

gw,gh,gc = goggle.shape

ex,ey,ew,eh = eyes[0][0],eyes[0][1],eyes[0][2],eyes[0][3]

for i in  range(0,gw):
    for j in range(0,gh):
        if(goggle[i,j][3] !=0):
            image[ey+i-22,ex+j-40] = goggle[i,j,1]
            
            
            
            
width = int(mustache.shape[1] * 25 / 100)
height = int(mustache.shape[0] * 25 / 100)
dim = (width, height)
mustache = cv2.resize(mustache, dim, interpolation = cv2.INTER_AREA)

mw,mh,mc = mustache.shape

nx,ny,nw,nh = nose[0][0],nose[0][1],nose[0][2],nose[0][3]
            
for i in  range(0,mw):
    for j in range(0,mh):
        if(mustache[i,j][3] !=0):
            image[ny+i+60,nx+j-5] = mustache[i,j,1]


print(ew)
plt.imshow(image)
```

```python
image1 = cv2.imread('Test/Before.png')
image1 =  cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)

goggle = cv2.imread('Train/glasses.png',-1)

mustache = cv2.imread('Train/mustache.png',-1)



eyes_detector = cv2.CascadeClassifier('Train/third-party/frontalEyes35x16.xml') 
eyes = eyes_detector.detectMultiScale(image1) 

nose_detector = cv2.CascadeClassifier('Train/third-party/Nose18x15.xml')
nose = nose_detector.detectMultiScale(image1,1.1,2)

plt.imshow(image1)

gw,gh,gc = goggle.shape

width_e = eyes[0][3]
height_e = eyes[0][2]
dim2 = (width_e, height_e)
goggle = cv2.resize(goggle, dim2, interpolation = cv2.INTER_AREA)

print(width_e,height_e,goggle.shape)




width_m = nose[0][3]
height_m= nose[0][2]
dim1 = (height_m, height_m)

mustache = cv2.resize(mustache, dim1, interpolation = cv2.INTER_AREA)

#print(mustache.shape)






for i in  range(0,height_e):
    for j in range(0,width_e):
        if(goggle[i,j][3] !=0):
            image1[eyes[0][1]+i,eyes[0][0]+j] = goggle[i,j,1]
            
            
            
            


mw,mh,mc = mustache.shape

nx,ny,nw,nh = nose[0][0],nose[0][1],nose[0][2],nose[0][3]
            
for i in  range(0,mw):
    for j in range(0,mh):
        if(mustache[i,j][3] !=0):
            image1[ny+i,nx+j] = mustache[i,j,1]


plt.imshow(image1)
```
