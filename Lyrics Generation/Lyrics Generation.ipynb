```python
import numpy as np
import pandas as pd
```

```python
def load_text(text_file):
    #with open(TextFile,encoding='utf-8') as f:  ## for hindi
    with open(text_file) as f:
        text = f.read()
    return text.lower()

```

```python
text = load_text('Apna Time Aayega.txt')
print(text)
```

```python
T = {}

def generateTable(text,k=4):
    for i in range(len(text)-k):
        x_text = text[i:i+k]
        y_text = text[i+k]
        if T.get(x_text) is None:
            T[x_text] = {}
            T[x_text][y_text] = 1
        else :
            if T[x_text].get(y_text) is None:
                T[x_text][y_text] = 1
            else:
                T[x_text][y_text] +=1
    for word in T.keys():
        s = sum(T[word].values())
        for i in T[word].keys():
            T[word][i] = T[word][i]/s
   
```

```python
generateTable(text)
#output = generateTable(text)
#print(output)
```

```python
def get_next_character(sentence,k=4):
    last_k_characters = sentence[-k:]
    if T.get(last_k_characters) is None:
        return " "
    possible_chars = list(T[last_k_characters].keys())
    possible_values = list(T[last_k_characters].values())
    
    #np.random.seed(11) # for consistent predictions
    sample_char = np.random.choice(possible_chars,p=possible_values)
    return sample_char
```

```python
def generateText(starting_sentence,maxLen=2000):
    sentence = starting_sentence
    for i in range(maxLen):
        next_character = get_next_character(sentence)
        sentence +=next_character
    return sentence
```

```python
print(generateText('Apna'))
```

```python
output_text = generateText('Apna')
print(output_text)
```

```python
#filename is complete path to file plus name and .txt extension
file = open('Lyrics_Generation_output.text', 'w')
file.write(output_text)
file.close()
```
